```python
import pandas as pd
import os

###Merging all 12 month in one dataset

df = pd.read_csv('./data/Sales_April_2019.csv')
all_months_data = pd.DataFrame()
files = [file for file in os.listdir('./data')]

for file in files:
  df = pd.read_csv('./data/'+file)
  all_months_data = pd.concat([all_months_data, df])

all_months_data = all_months_data.dropna()
all_months_data.to_csv('Sales_2019.csv', index=False)

```

```python
all_data = pd.read_csv('./Sales_2019.csv')
```

```python
# Best month for sales and how much was earned this month I guess loc or groupby using order date and first two digits as month and then calculate price

all_data = all_data[all_data['Order Date'].str[0:2]!='Or']



all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
all_data['Sales'] = all_data['Quantity Ordered']*all_data['Price Each']
```

```python
result = all_data.groupby('Month').sum()
```

```python
def get_city(city:str):
  return city.split(',')[1]


def get_state(state:str):
  return state.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + " " + get_state(x))

all_data.head()

```

```python
import matplotlib.pyplot as plt


months = range(1,13)

plt.bar(months, result['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD(million)')
plt.xlabel('Month number')
plt.show()
```

```python
result = all_data.groupby('City').sum()
result
```

```python
import matplotlib.pyplot as plt
city = [city for city, df in all_data.groupby('City')]

plt.bar(city, result['Sales'])
plt.xticks(city, rotation='vertical', size=8)
plt.ylabel('Sales in USD in milions')
plt.xlabel('City name')
plt.show()
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.grid()
plt.show()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ', '.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head(12)
```

```python
from itertools import combinations
from collections import Counter


count = Counter()

for row in df['Grouped']:
  row_list = row.split(', ')
  count.update(Counter(combinations(row_list, 2)))


count.most_common(10)
```

```python
product = all_data.groupby('Product')
quantity = product.sum('Quantity Ordered')['Quantity Ordered']

products = [product for product, df in product]

plt.bar(products, quantity)
plt.xticks(products, rotation='vertical', size=8)
plt.ylabel('Number Ordered')
plt.xlabel('Name')
plt.show()
```

```python
prices = all_data.groupby('Product').mean('Price Each')

fix, ax1 = plt.subplots()

ax2 = ax1.twinx()

ax1.bar(products, quantity)
ax2.plot(products, prices)


ax1.set_xlabel('Product Name')
ax1.set_xticklabels(products, rotation='vertical', size=8)
ax2.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Prices')

plt.show()
```
